# Limpieza de los datos

Empecé por armar algunos gráficos para darme una idea de qué metadata tiene cada subtítulo y con cuántos de los subtítulos se puede trabajar.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import pickle

datasets_path = "/Users/vale/Facultad/Tesis/data/datasets/"

with open(datasets_path + "detalles_subtitulos.p", 'rb') as f:
  subtitles_index = pickle.load(f, encoding='latin-1')
```

Resultado de contar por año (una vez por película, se cuentan MovieIDs diferentes)

```python
gr = subtitles_index[['MovieID','MovieYear']].groupby('MovieYear')
chart = gr.count().plot(kind='bar')
chart.xaxis.set_ticklabels(chart.xaxis.get_ticklabels(), rotation=45)
for i, label in enumerate(chart.xaxis.get_ticklabels()):
  if not i % 4 == 0:
    label.set_visible(False)
mpl.show()
```

Languages no es relevante porque son todos los idiomas en los que hay diálogos en la película (http://www.imdb.com/language/). El idioma de los subtítulos sale de SubLanguageID y están todos en inglés

```python
gr = subtitles_index[['MovieID','SubLanguageID']].groupby('SubLanguageID')
chart = gr.count().plot(kind='bar')
```
